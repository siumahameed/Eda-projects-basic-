```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df= pd.read_csv('data.csv', encoding="latin1")
df.head()
```

## Starting with basic data shape snd null values

```python
# we can see the shape of data by two ways. One is len(data) and another is data.shape
# More over we can see information about data by data.info()
df.info()
df.shape
```

```python
# We can check the null values in data by two ways. One is df.isnull().sum() and another is df.isnull().sum().sum()
df.isnull().sum().sum()
```

```python
# we can fill nul values by two ways. One is df.fillna(). We can do it by mean, median, mode, etc woth specific column
df['Profit'].fillna(df['Profit'].mean(), inplace=True)
# we can also drop the rows which contains null values by df.dropna()
```

## Analysis and graph of QUALITITVE DATA

```python
ship_mode = df['Ship Mode'].value_counts()
segment = df['Segment'].value_counts()
category = df['Category'].value_counts()
state= df['State'].value_counts() 
top_state= df['State'].value_counts().head(10)
region= df['Region'].value_counts()
sub_category= df['Sub-Category'].value_counts()
top_sub_category= df['Sub-Category'].value_counts().head(10)
```

```python
# We are ploting several bar charts for different features with qualititive data

plt.figure(figsize= (16,12), dpi=100, facecolor='white')

plt.subplot(2,2,1)
plt.bar(ship_mode.index, ship_mode.values)
plt.title('Frequency of Ship Mode', color='royalblue')
plt.xlabel('Ship Mode',fontdict=({ 'fontweight':'bold'}))
plt.ylabel('Count',fontdict=({ 'fontweight':'bold'}))

plt.subplot(2,2,2)
plt.bar(segment.index, segment.values)
plt.title('Frequency of Segment',color='royalblue')
plt.xlabel('Segment',fontdict=({ 'fontweight':'bold'}))
plt.ylabel('Count',fontdict=({ 'fontweight':'bold'}))

plt.subplot(2,2,3)
plt.bar(category.index, category.values)
plt.title('Frequency of Category',color='royalblue')
plt.xlabel('Category',fontdict=({ 'fontweight':'bold'}))
plt.ylabel('Count',fontdict=({ 'fontweight':'bold'})) 

plt.subplot(2,2,4)
plt.bar(region.index, region.values)
plt.title('Frequency of Region',color='royalblue')
plt.xlabel('Region',fontdict=({ 'fontweight':'bold'}))
plt.ylabel('Count',fontdict=({ 'fontweight':'bold'}))

                  

plt.show()
```

```python
plt.figure(figsize= (16,12), dpi=100, facecolor='white')
plt.subplot(2,2,1)
plt.title('Frequency of Ship Mode', color='black', fontdict=({ 'fontweight':'bold'}))
plt.pie(ship_mode.values, labels=ship_mode.index, autopct='%0.2f%%', startangle=90)
plt.legend(loc='best', bbox_to_anchor=(1, 1))

plt.subplot(2,2,2)
plt.title('Frequency of Segment', color='black', fontdict=({ 'fontweight':'bold'}))
plt.pie(segment.values, labels=segment.index, autopct='%1.1f%%', startangle=90)
plt.legend(loc='best', bbox_to_anchor=(1, 1))

plt.subplot(2,2,3)  
plt.title('Frequency of Category', color='black', fontdict=({ 'fontweight':'bold'}))
plt.pie(category.values, labels=category.index, autopct='%1.1f%%', startangle=90)
plt.legend(loc='best', bbox_to_anchor=(1, 1))

plt.subplot(2,2,4)  
plt.title('Frequency of Region', color='black', fontdict=({ 'fontweight':'bold'}))
plt.pie(region.values, labels=region.index, autopct='%1.1f%%', startangle=90)
plt.legend(loc='best', bbox_to_anchor=(1, 1))



plt.show()
```

```python
city_20 = df["City"].value_counts().head(10)
state_20 = df["State"].value_counts().head(10)
plt.figure(figsize= (16,12), dpi=100, facecolor='white', layout='constrained')

plt.subplot(2,1,1)
sns.barplot(x= state_20.index, y= state_20.values, palette='rainbow')

plt.subplot(2,1,2)
sns.barplot(x= city_20.index, y= city_20.values, palette='magma')

plt.show()
```

```python
plt.figure(figsize= (16,12), dpi=100, facecolor='white', layout='constrained')

plt.subplot(2,2,1)
plt.title("Relation between Ship Mode and Segment", color='royalblue', fontdict=({ 'fontweight':'bold'}))
sns.countplot(x='Ship Mode', data=df, hue='Segment',palette='rainbow')

plt.subplot(2,2,2)
plt.title("Relation between Ship Mode and Category", color='royalblue', fontdict=({ 'fontweight':'bold'}))
sns.countplot(x='Ship Mode', data=df, hue='Category',palette='magma')

plt.subplot(2,2,3)
plt.title("Relation between Ship Mode and Region", color='royalblue', fontdict=({ 'fontweight':'bold'}))
sns.countplot(x='Ship Mode', data=df, hue='Region',palette='viridis')


plt.subplot(2,2,4)
plt.title("Relation between Category and Region", color='royalblue', fontdict=({ 'fontweight':'bold'}))
sns.countplot(x='Category', data=df, hue='Region',palette='cubehelix')
plt.show()
```

```python
plt.figure(figsize= (16,12), dpi=100, facecolor='white', layout='constrained')

plt.subplot(2,2,1)
plt.title("Office Supplies Sub-Category Sale", color='royalblue', fontdict=({ 'fontweight':'bold', 'fontsize':14}))
sns.countplot(x='Category', data=df[df["Category"]=="Office Supplies"], hue='Sub-Category',palette='viridis')

plt.subplot(2,2,2)
plt.title("Furniture Sub-Category Sale", color='royalblue', fontdict=({ 'fontweight':'bold', 'fontsize':14}))
sns.countplot(x='Category', data=df[df["Category"]=="Furniture"], hue='Sub-Category',palette='rainbow')

plt.subplot(2,2,3)
plt.title("Technology Sub-Category Sale", color='royalblue', fontdict=({ 'fontweight':'bold', 'fontsize':14}))
sns.countplot(x='Category', data=df[df["Category"]=="Technology"], hue='Sub-Category',palette='magma')

plt.subplot(2,2,1)


plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Year'] = df['Order Date'].dt.year

df['Year'].value_counts()
sns.countplot(x='Year', data=df,hue='Category',palette='viridis')
```

```python
plt.figure(figsize= (12,5), dpi=100, facecolor='white', layout='constrained')

plt.subplot(1,2,1)
plt.title("Yearly Sales According to Category", color='royalblue', fontdict=({ 'fontweight':'bold', 'fontsize':12}))
sns.barplot(x="Category", y="Sales", data=df,hue="Year", palette='rainbow',estimator=np.sum)

plt.subplot(1,2,2)
plt.title("Yearly Profit According to Category", color='royalblue', fontdict=({ 'fontweight':'bold', 'fontsize':12}))
sns.barplot(x="Category", y="Profit", data=df, hue="Year", palette='rainbow',estimator=np.sum)

plt.show()
```
